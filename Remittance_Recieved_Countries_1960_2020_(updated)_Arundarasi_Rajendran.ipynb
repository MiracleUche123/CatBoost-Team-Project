* Transpose rows and columns

```python
df1 = df.rename(columns = {'Country Name':''})
df1.head()
```

```python
df2 = df1.set_index('')
df2.head()
```

```python
df3 = df2.drop(columns='Country Code')
df3.head()
```

```python
df4 = df3.transpose()
df4.head()
```

```python
df4.reset_index(inplace=True)
df4.head()
```

```python
df5 = df4.rename(columns = {'index':'Year'})
df5.head()
```

```python
df6 = df5.set_index('Year')
df6.head()
```

* Remove counties that do not have any data across all 4 datasets

```python
rows_with_nan = df6[df6.isnull().all(axis=1)]
```
