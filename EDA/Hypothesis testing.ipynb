H1: Countries with higher population are more likely to have high remittance received

```python
population = df['population']
remitt_received = df['remittance_received']
```

```python
corr1 = population.corr(remitt_received,method='pearson')
print("Correlation between Population and Remittance received is: ",corr1)
```

```python
#create a scatter plot
plt.figure(figsize=(12,5))
plt.scatter(population,remitt_received,c='green')
plt.xlabel('Population')
plt.ylabel('Remittance received')
plt.title("Population vs. Remittance received")
```

Observation

H5: Countries in certain regions are more likely to be considered as high income group

```python
df['new_income_group'].unique()
```

```python
income_mapping = {'Low income':1, 'Lower middle income':2,'Upper middle income':3,'High income':4}
```

```python
df['new_income_group']=df['new_income_group'].map(income_mapping)
```

```python
region = df['new_region']
avg_income = df.groupby('new_region')['new_income_group'].mean()
avg_income
```

```python
contingency_tab = pd.DataFrame(avg_income,region)
```

```python
contingency_tab
```

```python
from scipy.stats import chi2_contingency
chi2, p_value, _, _ = chi2_contingency(contingency_tab)
```

```python
print("Average Income level by region:")
print(avg_income)
print("")

print("Chi-Square Test Results:")
print("Chi-Square:", chi2)
print("p-value:", p_value)
```

```python
print("Minimum average income by region:", avg_income.min())
print("Maximum average income by region:", avg_income.max())
```
