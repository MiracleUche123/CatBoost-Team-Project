### Preprocessing

```python
paid_df.columns[1:4]
```

```python
def preprocess(df):
    df = df.copy()
    # Drop columns
    drop_cols = [str(col) for col in df.columns[1:4]]
    drop_cols.append('Unnamed: 67')
    
    ### remove country code, indicator name, indicator code and unnmaed:67
    df.drop(columns = drop_cols, inplace=True)

    # replace missing values with zero
    df.fillna(value=0, inplace =True)

    # Set index to Country Name
    df.set_index('Country Name', inplace=True)
    
    # Drop row Not Classified
    df = df.drop(index='Not classified', axis=0)
    
    # add total column
    df['Total'] = df.apply(lambda row: row.sum(), axis=1)

    # countries in the dataset with no data
    missing_countries = [x for x in df.index if df.loc[x]['Total']== 0]
    
    return df.T, missing_countries
```

```python
paid_df, null_paid = preprocess(paid_df)
received_df, null_received = preprocess(received_df)
gdp_df, null_gdp = preprocess(gdp_df)
unemp_ilo, null_unemp_ilo  = preprocess(unemp_ilo)
unemp_nat, null_unemp_nat = preprocess(unemp_nat)
```
